```python
document = [doc1,doc2,doc3,doc4,doc5,doc6,doc7,doc8,doc9,doc10]
```

```python
!pip install sentence-transformers
```

```python
vector_store = Chroma(
    embedding_function=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2"),
    persist_directory='my_chroma_db_store',
    collection_name='sample_data'
)
```

```python
#adding documents
vector_store.add_documents(document)
```

```python
vector_store.get(include=['embeddings','documents','metadatas'])
```

```python
#search document
vector_store.similarity_search(
    query='Who among these are bowler? ',
    k = 2 # fetch most 2 document
)
```

```python
#search with similarity_score => minimum means highest similarity
vector_store.similarity_search_with_score(
    query='who among these are bowlers? ',
    k = 2
)
```

```python
#metadata filtering
vector_store.similarity_search_with_score(
    query= "",
    filter={"team":"Chennai Super Kings"}
)
```

```python
#updating documents
updated_docs = Document(
    page_content='Virat Kohli is one is one of indian player',
    metadata={"team":"Royal Challengers Bangalore"}
)
vector_store.update_document(document_id='a82c0503-9cf7-4a80-8695-19f89d06ef06',document = updated_docs)
```
